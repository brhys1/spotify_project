```python
import pandas as pd

maya = pd.read_csv('maya_music_genres.csv')
rhys = pd.read_csv('rhys_spotify_songs.csv')
spotify_songs = pd.read_csv('spotify_songs.csv')
most_streamed = pd.read_csv('spotify-most-streamed.csv')
teagan = pd.read_csv('teagan_music_features.csv')
teagan2 = pd.read_csv('teagan_2music_features2.csv')
top_50 = pd.read_csv('top_50_2023.csv')

dfs = [maya, rhys, spotify_songs, most_streamed, teagan, teagan2, top_50]
```

```python
for df in dfs:
    print(df.shape[0])
    print(len(df.columns))
    print('--------------------------')
```

```python
dfs[0]['Track Name'] = dfs[0]['Track Name'].str.lower()
dfs[1]['track_name'] = dfs[1]['track_name'].str.lower()
dfs[2]['track_name'] = dfs[2]['track_name'].str.lower()
```

```python
dfs[0] = dfs[0].rename(columns={'Artist Name' : 'track_artist', 'Track Name' : 'track_name', 'Popularity' : 'track_popularity', 'duration_in min/ms' : 'duration_ms'})
```

```python
combined = pd.concat([dfs[1], dfs[2], dfs[0]], ignore_index=True)
```

```python
combined
```

```python
combined = combined.drop(columns=['Class', 'time_signature'])
```

```python
combined.shape[0]
```

```python
combined = combined.drop_duplicates(subset=['track_name', 'track_artist'], keep='first')  
combined.shape[0]
```
