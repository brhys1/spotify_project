```python
import pandas as pd

# Load the CSV
df = pd.read_csv("Spotify 2010 - 2019 Top 100.csv")

# Show the first few rows
df.head()
df.columns
# Convert columns to numeric (skip errors for now)
for col in ['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'pop']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df['dnce'].hist(bins=30)
plt.title("Danceability Distribution")
plt.xlabel("Danceability")
plt.ylabel("Count")
plt.show()
```

```python
df.boxplot(column='pop')
plt.title("Popularity Boxplot")
plt.show()
```

```python
sns.lmplot(x='dnce', y='pop', data=df)
plt.title("Danceability vs Popularity")
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_df = df[['bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'pop']]

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Audio Features")
plt.show()
```
